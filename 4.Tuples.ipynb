## 4. Tuples in Python

1. What is `Tuples?`
 * A `tuple` is a collection data type in Python used to store multiple values in a single variable.

### Key Characteristics:
 * __Ordered Collections__
 * __Immutable (cannot be changed after cration)__
 * __Allows duplicate values__
 * __Can sotre different data types__
 * __Faster than list__
 * __Written using round brackets (parentheses)__

```python
numbers = (10,20,30,40)
print(numbers)
print(type(numbers))
```

## 2. Creating Tuples
Method-1 : Using Parentheses

```python
fruits = ('apple', "banana","mango")
print(fruits,type(fruits))
```

### 3. Without Parentheses
Parentheses are optional

```python
colors = "red","green","blue"
print(colors,type(colors))
```

### 4. Nested Tuples
 A tuple inside another tuple.

```python
nested_tuple = (1,2,(3,4),(5,6))
print(nested_tuple)
print(type(nested_tuple))
```

### 5. Accessing Tuple Elements

Tuple elements are accessed using positive and negative indexing.

Positive Indexing

```python
fruits = ('apple', "banana","mango")
print(fruits,type(fruits))

print(fruits[0])
print(fruits[1])
print(fruits[2])
```

Negative Indexing

```python
fruits = ('apple', "banana","mango")
print(fruits,type(fruits))

print(fruits[-1])
print(fruits[-2])
print(fruits[-3])
```

## 6. Tuple Slicing

 Used to extract a protion of the tuple

syntax: 
  [start:end:step]

```python
numbers = (10,20,30,40,50)
print(numbers[1:4])  ## (20,30,40)
print(numbers[:3])  ### (10,20,30)
print(numbers[2:])  ### (30,40,50)
```
