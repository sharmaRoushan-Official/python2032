### 5. Python Set Data Type

### 1. Set In Python

 * A set in Python is a collection data type used to store multiple unique elements.

 * It is based on the mathematical concept of a set.

### Key Characteristics
1. Unordered
2. Unindexed
3. Mutable(can add/remove items)
4. Does not allow duplicate value
5. Elements must be immutable(hashable)



```python
#How to create empty set
s = {}
print(s,type(s))
```

```python
# create empty set 

s1 = set()
print(s1,type(s1))
```

```python
s = {1}
print(s,type(s))
```

2. Creating Sets 

Method 1: Using Curly Braces `{}`

```python
s = {1,2,3,4,5}
print(s,type(s))
```

Method 2: Uisng `set()` constructor

```python
s = set([1,1,2,3,6,65,54,6,6])
print(s,type(s))
```

### 3. Duplicate Elements 

sets automatically remove duplicates.

```python
s = {10,20,30,40,40}
print(s)
```

### 4. Data Types allowed in sets

Elements must be immutable(hashable)

int

float

string

tuple

boolean

frozenset

```python
s = {10,3.5,'Python',(1,2,3),True,False}
print(s)
```

#### Not Allowed

Mutable Datatypes:

list

dictionary

set

```python
s = {[1,2,3]}
print(s)
```

### 5. Accessing Set Elements

Since sets are unordered and unidexed, we cannot access elements using index.

```python
# Wrong
s = {10,20,30}
print(s[0])
```

```python
# Correct
s = {10,20,30}

for i in s:
    print(i)
```

### 6. Checking Membership

Use `in or not in` Operator

```python
s = {10,20,30}

print(10 in s)
print(10 not in s)
print(20 not in s)
print(30 in s)
```

### 7. Adding Elements

add()

Adds single element.

```python
s = {10,20}
s.add(30)
print(s)
```

```python
s.add(-100)
print(s)
```

## update()

Adds mutliple elements

```python
s = {10,20}

s.update([20,30,40,50,-8])
print(s)
```

### You Can add:

 * List
 * tuple
 * set

```python
s.update((10,20,-20))
print(s)
```

### 8 Removing Elements

remove()

Removes specific element

```python
s = {10,20,30}
s.remove(20)
print(s)
```

```python
s.remove(20)
print(s)
```

### discard()

Same as remove but no error if element is missing.

```python
s = {10,20,30}
s.discard(20)
print(s)
```

```python
s.discard(20)
print(s)
```

```python
s.discard(20)
print(s)
```

### clear()

Removes all elements at a time

```python
s = {1,2,3,4,5}

s.clear()
print(s)
```
