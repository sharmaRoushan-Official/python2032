### 5. Python Set Data Type

### 1. Set In Python

 * A set in Python is a collection data type used to store multiple unique elements.

 * It is based on the mathematical concept of a set.

### Key Characteristics
1. Unordered
2. Unindexed
3. Mutable(can add/remove items)
4. Does not allow duplicate value
5. Elements must be immutable(hashable)



```python
#How to create empty set
s = {}
print(s,type(s))
```

```python
# create empty set 

s1 = set()
print(s1,type(s1))
```

```python
s = {1}
print(s,type(s))
```

### 5. Accessing Set Elements

Since sets are unordered and unidexed, we cannot access elements using index.

```python
# Wrong
s = {10,20,30}
print(s[0])
```

```python
# Correct
s = {10,20,30}

for i in s:
    print(i)
```

### 6. Checking Membership

Use `in or not in` Operator

```python
s = {10,20,30}

print(10 in s)
print(10 not in s)
print(20 not in s)
print(30 in s)
```

### 10. Set Operations ()

Python sets support mathematical operations.



1. Union

Combines elemnets of both sets.

Operator:

|   <- pipe Operator

```python
a = {1,2,3}
b = {3,4,5}

print(a | b)
```

```python
# 2nd method

print(a.union(b))
```

### 2. Intersection 

Comman Elements

operator - &

```python
a = {1,2,3}
b = {3,4,5}

print(a & b)
```

```python
# 2nd method 

print(a.intersection(b))
```

### 3.Difference

Elements present in first but not second 

operator: -

```python
a = {1,2,3}
b = {2,3,4}

print(a-b)  # 1
print(b-a)  # 4
```

```python
# 2nd method 

print(a.difference(b))
print(b.difference(a))
```

### 4. Symmetric Difference

Elemnets in either set but not in both.

operator: ^ [caret]


```python
a = {1,2,3}
b = {3,4,5}

print(a ^ b )
```

```python
a = {1,2,3}
b = {2,3,4}

print(a ^ b)
```

```python
# 2nd method 

print(a.symmetric_difference(b))
```

### 11. Set Comparison Operations

subset

Check if set is subset

```python
a = {1,2}  # a is subeset of b
b = {1,2,3}   # b is superset of a

print(a.issubset(b))
print(a.issuperset(b)) # false


print(b.issubset(a))  # false
print(b.issuperset(a))
```

### disjoint

Checks if no commen elements

```python
a = {1,2}
b = {3,4}

print(a.isdisjoint(b))
```

```python
a = {1,2,3}
b = {3,4}

print(a.isdisjoint(b))
```

### 14. Performance Advantage of Sets
---
Sets use hash tables, so operation are very fast.


| Time Complexity Class | Description | Example |
|----------------------|-------------|---------|
| **O(1) – Constant Time** | The algorithm takes the same amount of time regardless of input size. | Accessing an array index, basic arithmetic operations |
| **O(log n) – Logarithmic Time** | Runtime increases slowly as the input size increases, usually due to dividing the problem into smaller parts. | Binary Search |
| **O(n) – Linear Time** | Execution time grows proportionally with the input size. | Traversing an array |
| **O(n log n) – Linearithmic Time** | Combination of linear and logarithmic growth, commonly seen in efficient sorting algorithms. | Merge Sort, Quick Sort (average case), Heap Sort |
