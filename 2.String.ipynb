# Indexing of a string

```python
msg = "Hello world"
print(msg)
```

```python
# positive Indexing
print(msg[0])
print(msg[1])
print(msg[2])
print(msg[3])
print(msg[4])
print(msg[5])
print(msg[6])
print(msg[7])
print(msg[8])
print(msg[9])
print(msg[10])
```

```python
# Negative Indexing

print(msg[-1])
print(msg[-2])
print(msg[-3])
print(msg[-4])
print(msg[-5])
print(msg[-6])
print(msg[-7])
print(msg[-8])
print(msg[-9])
print(msg[-10])
print(msg[-11])
```

```python
# to count total length of string
print(len(msg))
```

```python
msg = "Hello world"
print(msg)
```

# Slicing of string

```python
msg1 = msg[0:5]  # n-1 = 5-1 = 4
print(msg1)
```

```python
msg2 = msg[6:11]
print(msg2)
```

```python
msg1 = msg[:5]  # n-1 = 5-1 = 4
print(msg1)
```

```python
msg2 = msg[6:]
print(msg2)
```

## Escape Sequence

```python
print("Hello world\nHow many people are living on the earth?") # \n - new line
```

```python
print("///")
```

```python
print("\\\\\\")
```

```python
print("Hello world\tHow many people are living on the earth?") # \n - new line
```

| Escape Sequence | Description               | Example                    | Output           |
| --------------- | ------------------------- | -------------------------- | ---------------- |
| `\\`            | Backslash                 | `print("Hello\\World")`    | Hello\World      |
| `\'`            | Single Quote              | `print('It\'s Python')`    | It's Python      |
| `\"`            | Double Quote              | `print("She said \"Hi\"")` | She said "Hi"    |
| `\n`            | New Line                  | `print("Hello\nWorld")`    | Hello <br> World |
| `\t`            | Horizontal Tab            | `print("Hello\tWorld")`    | Hello World      |
| `\r`            | Carriage Return           | `print("Hello\rWorld")`    | World            |
| `\b`            | Backspace                 | `print("Hell\b o")`        | Hel o            |
| `\f`            | Form Feed                 | `print("Hello\fWorld")`    | Hello World      |
| `\v`            | Vertical Tab              | `print("Hello\vWorld")`    | Hello World      |
| `\a`            | Bell / Alert Sound        | `print("\a")`              | System beep      |
| `\ooo`          | Octal Value               | `print("\110")`            | H                |
| `\xhh`          | Hexadecimal Value         | `print("\x48")`            | H                |
| `\N{name}`      | Unicode Character by Name | `print("\N{HEART SUIT}")`  | ♥                |
| `\uXXXX`        | Unicode (16-bit)          | `print("\u2764")`          | ❤                |
| `\UXXXXXXXX`    | Unicode (32-bit)          | `print("\U0001F600")`      | 😀               |
