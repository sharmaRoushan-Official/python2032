# 3. Lists in Python

 * __Lists are ordered__
 * __Lists can contains any arbitrary objects.__
 * __List elements can be accessed by index.__
 * __Lists can be nested to arbitrary depth.__
 * __Lists are mutable.__
 * __Lists are dynamic.__

## 1. What is a List?

 * __A `List` is a built-in data type in python.__
 * __It is used to store `multiple values in a single variable.`__
 * __Lists are:__
   * __Ordered__
   * __Mutable(changeable)__
   * __Allow duplicate values__
   * __Can store different data types.__

```python
# empty list
num = []
print(num)
print(type(num))
```

```python
# list with data 
numbers = [10,20,30,40]
print(numbers)
```

# Explanation:
 * `numbers` is a list.
 * Elements are stored inside square brackets `[]`.
 * Order is maintained (10 is first, 20 is second, 30 is third and 40 is foruth)]

```python
mixed_list = [1,'Python',3.5,True,False]
print(mixed_list)
```

```python
# Positive Indexing
print(mixed_list[0])
print(mixed_list[1])
print(mixed_list[2])
print(mixed_list[3])
print(mixed_list[4])
```

```python
# Negative Indexing
print(mixed_list[-1])
print(mixed_list[-2])
print(mixed_list[-3])
print(mixed_list[-4])
print(mixed_list[-5])
```

```python
mixed_list = [1,'Python',3.5,True,False]
print(mixed_list)

print(mixed_list[0],type(mixed_list[0]))
print(mixed_list[1],type(mixed_list[1]))
print(mixed_list[2],type(mixed_list[2]))
print(mixed_list[3],type(mixed_list[3]))
print(mixed_list[4],type(mixed_list[4]))
```
