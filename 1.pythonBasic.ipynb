### type()

```python
# 1.String
print("hello world")
print(type("hello world"))
print('Hello world')
print(type('Hello world'))
print('1000')
print(type('1000'))
print('123.56')
print(type('123.56'))
print('True')
print(type('True'))
print("456.25")
print(type("456.25"))


# 2. Integer
print(100)
print(type(100))
print(12000)
print(type(12000))
print(450)
print(type(450))
print(78)
print(type(78))


# 3. Float
print(123.25)
print(type(123.25))
print(100.0)
print(type(100.0))
print(78.23)
print(type(78.23))
print(789.56)
print(type(789.56))


# 4. Boolean
print(True)
print(type(True))
print(False)
print(type(False))
```
